# Import Library

```python
from sklearn.model_selection import GridSearchCV, cross_val_score, cross_validate, KFold
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, confusion_matrix, roc_auc_score, recall_score, f1_score, precision_score, accuracy_score
from sklearn.svm import SVC
import joblib
from google.colab import files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Load Data

```python
!git clone https://github.com/Hafiyyanrevi/Drug-Side-Effect-Prediction-GSA-SVM.git data_TA
```

```python
X_train = pd.read_csv('data_TA/results/PreProcessing/X_train_smilesdataset.csv')
y_train = pd.read_csv('data_TA/results/PreProcessing/y_train_smilesdataset.csv')
```

```python
X_test = pd.read_csv('data_TA/results/PreProcessing/X_test_smilesdataset.csv')
y_test = pd.read_csv('data_TA/results/PreProcessing/y_test_smilesdataset.csv')
```

**Load theBest_features pickle file of feature selection result**

```python
with open('data_TA/results/FeatureSelection/linear/theBest_features_linear.pkl', 'rb') as pickle_file:
    theBest_features_linear = joblib.load(pickle_file)
```

```python
X_train_selected_linear = X_train.loc[:, theBest_features_linear]
```

# Hyperparameter Tuning

**Hyperparameter search space**

```python
param_grid = {
    'C': [0.1, 1, 10, 100, 1000],
    'gamma': ['scale', 'auto'],
    'kernel': ['linear'],
    'degree': [1, 2, 3, 4, 5]
}
```

```python
svc = SVC()
```

**Create a GridSearchCV object**

```python
linear = GridSearchCV(SVC(kernel='linear'), param_grid, refit=True, cv=10, scoring='f1', verbose= 3)
```

**Hyperparameter tuning on data with selected features**

```python
linear.fit(X_train_selected_linear, y_train)
```

**Default parameter**

```python
default_svc = SVC()
default_params = default_svc.get_params()
```

```python
from pprint import pprint

print("Default parameters for SVC: ")
pprint(default_params)
```

```python
default = {'C': 1, 'degree': 3, 'gamma': 'scale', 'kernel': 'linear'}
```

**Display the Mean Test Score of all hyperparameter tuning result parameters**

```python
results_linear = linear.cv_results_

joblib.dump(results_linear, "results_linear.pkl")

print("Results:")
for mean_score, params in zip(results_linear['mean_test_score'], results_linear['params']):
    print(f"Mean Test Score: {mean_score:.5f}, Parameter: {params}")
```

**Comparison of best parameters with default parameters**

```python
def find_mean_score(best_params, cv_results):
    for mean_score, params in zip(cv_results['mean_test_score'], cv_results['params']):
        if params == best_params:
            return mean_score
    return None
```

```python
def find_mean_default(default_params, cv_results):
    for mean_score, params in zip(cv_results['mean_test_score'], cv_results['params']):
        if params == default_params:
            return mean_score
    return None
```

```python
best_params_linear = linear.best_params_
mean_score_linear = find_mean_score(best_params_linear, results_linear)
mean_accuracy_default = find_mean_default(default, results_linear)
```

```python
print("\n||------------------------ linear ------------------------")
print("|| Best parameters found for linear kernel: ", best_params_linear)
print("|| Mean Test Score Best parameters: ", mean_score_linear)
print("|| Mean Accuarcy of Default parameters: ", mean_accuracy_default)
```

```python
joblib.dump(mean_score_linear, "mean_score_linear.pkl")
joblib.dump(mean_accuracy_default, "mean_accuracy_default.pkl")
```

# Model Performance Evaluation

**Saving the best model**

```python
best_model_linear = linear.best_estimator_
```

**Showing the best model**

```python
best_model_linear

linear_score = cross_val_score(best_model_linear, X_train_selected_linear, y_train, cv=10, scoring='f1')
linear_score = np.average(linear_score)
```

```python
print('Score: ',linear_score)
```

```python
selected_features_linear_indices = X_test.columns.get_indexer(theBest_features_linear)
```

**Evaluate the best model on test data**

```python
test_accuracy_linear = best_model_linear.score(X_test.iloc[:, selected_features_linear_indices], y_test)
```

```python
print("Test set accuracy of best model for linear kernel: ", test_accuracy_linear)
```
